## 1.	Đọc dữ liệu từ file ‘mpg.csv’. Hiển thị 10 dòng đầu tiên.

```python
import pandas as pd
import numpy as np

df = pd.read_csv('/content/sample_data/mpg.csv');
df.head(10)
```

## 2.	Tìm hiểu dữ liệu:<br>
a.	Trong tập dữ liệu, thuộc tính nào định tính, thuộc tính nào định lượng<br>
<ul><B>Thuộc tính định tính:</B>
<li>model_year</li>
<li>origin</li>
<li>name</li>
</ul>
<ul><B>Thuộc tính định lượng:</B>
<li>mpg</li>
<li>cylinders</li>
<li>weight</li>
<li>acceleration</li>
</ul>

c.	Cho biết miền giá trị và miền phân vị của thuộc tính mpg

```python
print('Miền giá trị:',df.mpg.max() - df.mpg.min())
print('Miền phân vị:',np.percentile(df.mpg, 75) - np.percentile(df.mpg, 25))
```

b.	Hãy cho biết trung bình và trung vị của các thuộc tính: cylinders, horsepower, weight

```python
df.describe()[['cylinders', 'horsepower', 'weight']].loc[['mean', '50%']]
```

3.	Hãy so sánh mức tiêu hao nhiên liệu trung bình của các dòng xe của Mỹ, Nhật, và Châu Âu

```python
df[['origin', 'mpg']].groupby(['origin']).mpg.mean()
```

4.	Hãy so sánh công suất trung bình của các dòng xe của Mỹ, Nhật, và Châu Âu

```python
df[['origin', 'horsepower']].groupby(['origin']).mean().reset_index('origin')
```

5.	Tính tỷ lệ của các dòng xe Mỹ, Nhật, Châu Âu trong bộ dữ liệu.

```python
print('USE: {:.2f}%'.format(df[df.origin == 'usa'].count().origin*100 /  df.count().origin))
```

```python
print('JAPAN: {:.2f}%'.format(df[df.origin == 'japan'].count().origin*100 /  df.count().origin))
```

```python
print('EUROPE: {:.2f}%'.format(df[df.origin == 'europe'].count().origin*100 /  df.count().origin))
```

6.	Vẽ đồ thị histogram trọng lượng của các xe.

```python
import matplotlib.pyplot as plt
font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'darkred','size':15}
plt.title('Trọng lượng của các xe', fontdict=font1)
plt.xlabel('Trọng lượng', fontdict=font2)
plt.ylabel('Số lượng', fontdict=font2)
plt.hist(df.weight, rwidth = 0.97)
plt.show()
```

7.	Vẽ đồ thị boxplot so sánh công suất của các dòng xe thập niên 70 và thập niên 80

```python
import seaborn as sns
import matplotlib.pyplot as plt
decade70 = df[df.model_year.between(70, 79)].horsepower.values
decade80 = df[df.model_year.between(80, 89)].horsepower.values

font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'darkred','size':15}

plt.figure(2, figsize=(9, 6))
sns.boxplot(decade70, color='red')
plt.title('Công suất dòng xe thập niên 70', fontdict=font1)
plt.xlabel('horsepower', fontdict=font2)
plt.show()

plt.figure(2, figsize=(9, 6))
sns.boxplot(decade80, color='cyan')
plt.title('Công suất dòng xe thập niên 80', fontdict=font1)
plt.xlabel('horsepower', fontdict=font2)
plt.show()
```

8.	Bằng cách chọn biểu đồ phù hợp bạn hãy khảo sát mức tiêu hao nhiên liệu (mpg) theo các thuộc tính như


a.	Mức tiêu hao nhiên liệu và số lượng xi lanh

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(20, 10))
plt.plot(df.mpg.values, label='mpg')
plt.plot(df.cylinders.values, label='cylinders')
plt.legend()
plt.show()
```

b.	Mức tiêu hao nhiên liệu và công suất theo mã lực

```python
plt.figure(figsize=(20, 10))
plt.plot(df.mpg.values, label='mpg')
plt.plot(df.horsepower.values, label='horsepower')
plt.legend()
plt.show()
```

c.	Mức tiêu hao nhiên liệu và gia tốc

```python
plt.figure(figsize=(20, 10))
plt.plot(df.mpg.values, label='mpg')
plt.plot(df.acceleration.values, label='acceleration')
plt.legend()
plt.show()
```

d.	Mức tiêu hao nhiên liệu theo năm sản xuất

```python
import plotly.express as px

px.bar(df[['mpg', 'model_year']], x='model_year', y='mpg')
```
