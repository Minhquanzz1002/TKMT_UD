# Bài Tập

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import norm
```

### Bài 1: Tạo quần thể mô phỏng

Bạn hãy tạo một mẫu gồm 10000 phần tử mô phỏng chiều cao của nam thanh niên có giá trị từ 120cm - 200cm. Bạn hãy lưu kết quả vào biến **POP**.

Tính:

a. Chiều cao trung bình (kỳ vọng) của quần thể và độ lệch chuẩn về chiều cao của quần thể.

b. Tính tỷ lệ người cao trong quần thể, biết rằng thanh niên có chiều cao từ 180 trở lên được gọi là cao.

c. Vẽ histogram về chiều cao của quần thể. Theo bạn quần thể có phân phối chuẩn hay không?

d. Thử vẽ histogram và đồ thị hàm mật độ của phân phối chuẩn sử dụng tham số loc và scale bằng với kỳ vọng và độ lệch chuẩn của quần thể.

#### Chiều cao trung bình (kỳ vọng) của quần thể và độ lệch chuẩn về chiều cao của quần thể.

```python
# hàm tái thiết lặp thí nghiệm
np.random.seed(25)
```

```python
N = 10000
POP = np.random.randint(120, 200, N)
```

```python
POP_mean = POP.mean()
POP_std = POP.std()
```

```python
print('Chiều cao trung bình của quần thể {}'.format(POP_mean.round(4)))
print('Độ lệch chuẩn về chiều cao của quần thể {}'.format(POP_std.round(4)))
```

#### Tính tỷ lệ người cao trong quần thể, biết rằng thanh niên có chiều cao từ 180 trở lên được gọi là cao.

```python
tyle_nguoicao = (POP>=180).sum() *100 / N
```

```python
print('Tỷ lệ người cao từ 180 trở lên {}%'.format(tyle_nguoicao.round(2)))
```

#### Vẽ histogram về chiều cao của quần thể. Theo bạn quần thể có phân phối chuẩn hay không?

##### Vẽ bằng pyplot

```python
plt.hist(POP, edgecolor='white', bins=30)
plt.xlabel('Chiều cao')
plt.ylabel('Số lượng')
plt.show()
```

##### Vẽ bằng seaborn

```python
sns.distplot(POP, kde=False, hist_kws=dict(edgecolor='white'), bins=30, color='blue')
plt.xlabel('Chiều cao')
plt.ylabel('Số lượng')
plt.xlim(120, 180)
plt.show()
```

#### Thử vẽ histogram và đồ thị hàm mật độ của phân phối chuẩn sử dụng tham số loc và scale bằng với kỳ vọng và độ lệch chuẩn của quần thể.

```python
x = np.linspace(POP_mean - 3*POP_std, POP_mean +3*POP_std, 100)
y = norm.pdf(x, loc= POP_mean, scale= POP_std)

plt.title('Phân phối chuẩn $\mu={}, \sigma={}$'.format(POP_mean.round(4), POP_std.round(4)))
data = norm.rvs(POP_mean, POP_std, size=10000)
plt.hist(data, edgecolor='w', density=True)
plt.plot(x, y)
plt.show()
```

### Bài 4: Khảo sát phân phối của trung bình mẫu

a. Thực hiện bài 2 100 lần, mỗi lần bạn tính được trung bình mẫu. Vẽ đồ thị histogram cho 100 trung bình mẫu bạn tính được.

b. Thực hiện bài 3 100 lần, mỗi lần bạn tính được trung bình mẫu. Vẽ đồ thị histogram cho 100 trung bình mẫu bạn tính được.

c. Vẽ histogram của hai câu a, b trong cùng một hình để so sánh. Theo bạn hình dạng của histogram thay đổi nói lên điều gì?

Gợi ý: tạo một danh sách rỗng để chứa các trung bình mẫu tính được, sử dụng hàm append() để thêm giá trị trung bình vào danh sách sau mỗi lần tính.

#### Thực hiện bài 2 100 lần, mỗi lần bạn tính được trung bình mẫu. Vẽ đồ thị histogram cho 100 trung bình mẫu bạn tính được.

```python
# khởi tạo mảng rỗng
arr_trungbinhmau_20 = []   
```

```python
for i in range(100):
    trungbinhmau = np.random.choice(POP, size=20).mean()
    arr_trungbinhmau_20.append(trungbinhmau)
```

```python
plt.hist(arr_trungbinhmau_20, edgecolor='w', bins=20)
plt.xlabel('Trung bình chiều cao')
plt.ylabel('Số lượng')
plt.title('Histogram của trung bình mẫu cỡ 20')
plt.show()
```

```python
sns.distplot(arr_trungbinhmau_20, kde=False, hist_kws=dict(edgecolor='w'), bins=20)
plt.xlabel('Trung bình chiều cao')
plt.ylabel('Số lượng')
plt.title('Histogram của trung bình mẫu cỡ 20')
plt.show()
```

#### Thực hiện bài 3 100 lần, mỗi lần bạn tính được trung bình mẫu. Vẽ đồ thị histogram cho 100 trung bình mẫu bạn tính được.

```python
arr_trungbinhmau_100 = []
```

```python
for i in range(100):
    trungbinhmau_100 = np.random.choice(POP, size=100).mean()
    arr_trungbinhmau_100.append(trungbinhmau_100)
```

```python
plt.hist(arr_trungbinhmau_100, edgecolor='w', bins=20)
plt.xlabel('Trung bình chiều cao')
plt.ylabel('Số lượng')
plt.title('Histogram của trung bình mẫu cỡ 100')
plt.show()
```

```python
sns.distplot(arr_trungbinhmau_100, kde=False, hist_kws=dict(edgecolor='w'), bins=20)
plt.xlabel('Trung bình chiều cao')
plt.ylabel('Số lượng')
plt.title('Histogram của trung bình mẫu cỡ 100')
plt.show()
```

#### Vẽ histogram của hai câu a, b trong cùng một hình để so sánh. Theo bạn hình dạng của histogram thay đổi nói lên điều gì?

```python
plt.hist(arr_trungbinhmau_20, edgecolor='w', bins=30, label='Cỡ mẫu 20')
plt.hist(arr_trungbinhmau_100, edgecolor='w', bins=30, label='Cỡ mẫu 100')
plt.legend()
plt.xlabel('Trung bình chiều cao')
plt.ylabel('Số lượng')
plt.show()
```

**<u>Nhận xét</u>**
- Histogram của đồ thị có mẫu cỡ 100 gần giống với phân phối chuẩn
- Histogram của đồ thị có cỡ mẫu 20 không giống với phân phối chuẩn

### Bài 5: Khảo sát tỷ lệ mẫu

a. Bạn hãy mô phỏng 1000 lần lấy mẫu có kích cỡ 100. Vẽ histogram tỷ lệ người cao của các mẫu

b. Bạn hãy mô phỏng 1000 lần lấy mẫu có kích cỡ 500. Vẽ histogram tỷ lệ người cao của các mẫu

#### Bạn hãy mô phỏng 1000 lần lấy mẫu có kích cỡ 100. Vẽ histogram tỷ lệ người cao của các mẫu

```python
arr_tylenguoicao_co100 = []
```

```python
for i in range(1000):
    tylenguoicao_co100 = (np.random.choice(POP, size=100)>=180).sum() / 100
    arr_tylenguoicao_co100.append(tylenguoicao_co100)
```

```python
plt.hist(arr_tylenguoicao_co100, edgecolor='w', bins=35)
plt.xlabel('Tỷ lệ người cao')
plt.ylabel('Số lượng')
plt.title('Histogram tỷ lệ người cao mẫu cỡ 100 mô phẩn 1000 lần')
plt.show()
```

#### Bạn hãy mô phỏng 1000 lần lấy mẫu có kích cỡ 500. Vẽ histogram tỷ lệ người cao của các mẫu

```python
arr_tylenguoicao_co500 = []
```

```python
for i in range(1000):
    tylenguoicao_co500 = (np.random.choice(POP, size=500)).sum() / 500
    arr_tylenguoicao_co500.append(tylenguoicao_co500)
```

```python
sns.distplot(arr_tylenguoicao_co500, kde=False, hist_kws=dict(edgecolor='w'), bins=35)
plt.xlabel('Tỷ lệ người cao')
plt.ylabel('Số lượng')
plt.title('Histogram tỷ lệ người cao mẫu cỡ 500 mô phẩn 1000 lần')
plt.show()
```

### Bài 6: Khảo sát phương sai mẫu

Bạn hãy mô phỏng 1000 lần lấy mẫu có kích cỡ 100. Vẽ histogram đại lượng phương sai của các mẫu (Phương sai bằng bình phương độ lệch chuẩn)

```python
arr_phuongsai_co100 = []
```

```python
for i in range(1000):
    phuongsai_co100 = np.var(np.random.choice(POP, size=100))
    arr_phuongsai_co100.append(phuongsai_co100)
```

```python
sns.distplot(arr_phuongsai_co100, kde=False, hist_kws=dict(edgecolor='w'), bins=40)
plt.xlabel('Phương sai')
plt.ylabel('Số lượng')
plt.title('Histogram đại lượng phương sai mẫu cỡ 100 mô phổng 1000 lần')
plt.show()
```
